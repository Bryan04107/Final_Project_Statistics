```python
import pandas as pd
df = pd.read_csv('output2.csv')

# For Non Toxic and Toxic
kda_not_toxic = df[df['label'] == 0]
kda_toxic = df[df['label'] == 1]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import statistics
import numpy as np

# Box and Whisker for Non Toxic
plt.figure(figsize=(20, 2))
sns.boxplot(x='kda', data=kda_not_toxic)
plt.xlabel('KDA for Non Toxic Messages')
plt.title('Box and Whisker Plot for KDA of Non-Toxic Messages')
plt.xlim(-0.5, 7.5)
plt.show()

# Mean, Standard Dev, Lower Quart, Upper Quart, Inter Quart for Non Toxic
lq, uq = np.percentile(kda_not_toxic['kda'], [25 ,75])
iqr = uq - lq

tMin = lq - 1.5 * iqr
Min = np.inf
for x in kda_not_toxic['kda']:
    if tMin < x < Min:
        Min = x

tMax = uq + 1.5 * iqr
Max = -np.inf
for y in kda_not_toxic['kda']:
    if tMax > y > Max:
        Max = y

print("Mean: {:.2f}".format(statistics.mean(kda_not_toxic['kda'])))
print("Standard Deviation: {:.2f}".format(statistics.stdev(kda_not_toxic['kda'])))
print("Min: {:.2f}".format(Min))
print("Max: {:.2f}".format(Max))
print("Median: {:.2f}".format(statistics.median(kda_not_toxic['kda'])))
print("Lower Quartile: {:.2f}".format(lq))
print("Upper Quartile: {:.2f}".format(uq))
print("Inter Quartile Range: {:.2f}".format(iqr))

# Box and Whisker for Toxic
plt.figure(figsize=(20, 2))
sns.boxplot(x='kda', data=kda_toxic)
plt.xlabel('KDA for Toxic Messages')
plt.title('Box and Whisker Plot for KDA of Toxic Messages')
plt.xlim(-0.5, 7.5)
plt.show()

# Mean, Standard Dev, Lower Quart, Upper Quart, Inter Quart for Toxic
min2, max2 = np.percentile(kda_toxic['kda'], [0 ,100])
lq2, uq2 = np.percentile(kda_toxic['kda'], [25 ,75])
iqr2 = uq2 - lq2

tMin2 = lq2 - 1.5 * iqr2
Min2 = np.inf
for x in kda_toxic['kda']:
    if tMin2 < x < Min2:
        Min2 = x

tMax2 = uq2 + 1.5 * iqr2
Max2 = -np.inf
for y in kda_toxic['kda']:
    if tMax2 > y > Max2:
        Max2 = y

print("Mean: {:.2f}".format(statistics.mean(kda_toxic['kda'])))
print("Standard Deviation: {:.2f}".format(statistics.stdev(kda_toxic['kda'])))
print("Min: ", Min2)
print("Max: ", Max2)
print("Median: {:.2f}".format(statistics.median(kda_toxic['kda'])))
print("Lower Quartile: {:.2f}".format(lq2))
print("Upper Quartile: {:.2f}".format(uq2))
print("Inter Quartile Range: {:.2f}".format(iqr2))
```

```python
# Normal Distribution for Non Toxic
plt.figure(figsize=(8, 3))
sns.histplot(x='kda', data=kda_not_toxic, element='step', kde=True)
plt.xlabel('KDA for Non-Toxic Messages')
plt.title('Normal Distribution Graph for KDA of Non-Toxic Messages')
plt.xlim(0, 10)
plt.ylim(0, 600)
plt.show()

# Normal Distribution for Toxic
plt.figure(figsize=(8, 3))
sns.histplot(x='kda', data=kda_toxic, element='step', kde=True)
plt.xlabel('KDA for Toxic Messages')
plt.title('Normal Distribution Graph for KDA of Toxic Messages')
plt.xlim(0, 10)
plt.ylim(0, 600)
plt.show()
```

```python
import scipy.stats

t_statistic, p_value_t = scipy.stats.ttest_ind(kda_not_toxic['kda'].tail(500), kda_toxic['kda'].tail(500))
correlation_coefficient_kda, p_value_kda = scipy.stats.pearsonr(kda_not_toxic['kda'].tail(500), kda_toxic['kda'].tail(500))

print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value (T-test): {p_value_t:.4f}")

print(f"Pearson Correlation Coefficient (KDA): {correlation_coefficient_kda:.4f}")
print(f"P-value (Pearson Correlation): {p_value_kda:.4f}")
```
